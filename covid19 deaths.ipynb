# Covid-19 death prediction

```python
import numpy as np
import pandas as pd
import plotly.express as px
```

### Loading dataset

```python
df = pd.read_excel('corona.xlsx')
df.head()
```

### Eliminating unwanted column

```python
df2=df
df2 = df2.drop("Unnamed: 0", axis=1)
df2
```

### Data Cleaning

```python
df2.isnull().sum()
```

### Plotting daily confirmed cases 

```python
fig = px.bar(df2, x='Date_YMD', y='Daily Confirmed')
fig.show()
```

# Covid-19 death rate analysis
- Visualizing the death rate due to Covid 19 

```python
cases = df["Daily Confirmed"].sum()
deceased = df["Daily Deceased"].sum()

labels = ["Confirmed", "Deceased"]
values = [cases, deceased]

fig = px.pie(df, values=values, 
             names=labels, 
             title='Daily Confirmed Cases vs Daily Deaths', hole=0.5)
fig.show()
```

### Calculating the death rates of covid-19

```python
death_rate = (df["Daily Deceased"].sum() / df["Daily Confirmed"].sum()) * 100

print(death_rate,"% per day")
```

### Visualizing daily deaths by covid-19
- We can see a high number of deaths during the high wave of covid-19 cases.

```python
fig = px.bar(df, x='Date_YMD', y='Daily Deceased')
fig.show()
```

# Covid-19 Deaths Prediction Model
- Here I will be using the AutoTS library, which is one of the best Automatic Machine Learning libraries for Time Series Analysis.

```python
from autots import AutoTS
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(df, date_col="Date_YMD", value_col='Daily Deceased', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)

```
